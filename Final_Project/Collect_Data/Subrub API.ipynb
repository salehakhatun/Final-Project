```python
##suburb Domain ID

# Configure gmaps with API key
import pandas as pd
import requests

# Import API key
from api_keys import g_key
import pprint as pprint
# loading the dataframe with all the information that saved in previous task
source_df=pd.read_csv('../VIC.csv')
source_df

start=392
end=892
for i in range(start,end):

 
   # set up a parameters dictionary
    params = {
        "searchLevel": "suburb",
        "suburb": source_df['Suburb'].iloc[i],
        "state": 'VIC',
  
        "api_key": g_key
    }
 
    # base url
    base_url = "https://api.domain.com.au/v1/addressLocators"
 
    # run a request using our params dictionary
    response = requests.get(base_url, params=params).json()
   
    # extract results
    #results = response['results']
    
    try:
            #append to hotel data frame
      #  pprint(results)
      print(response[0]['ids'][0]['id'])
      source_df['Domain_id'].iloc[i]=response[0]['ids'][0]['id']
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")

    


```

```python
# Configure gmaps with API key
import pandas as pd
import requests
import json

# Import API key
from api_keys import g_key
from pprint import pprint 
# loading the dataframe with all the information that saved in previous task
source_df=pd.read_csv('../VIC.csv')
start=2100
end=2310
for index in range(start,end):
    suburb=source_df.iloc[[index]]
    print(source_df['Domain_id'][index])
    
   # set up a parameters dictionary
    params = {
        "propertyCategory":"house",
        "suburbId": source_df['Domain_id'][index],
        "state": 'VIC',
        "chronologicalSpan":3,
        "tPlusFrom":1,
        "tPlusTo":1,
  
        "api_key": g_key
    }
 
    # base url
    base_url = "https://api.domain.com.au/v1/suburbPerformanceStatistics"
 
    
   
    # extract results
    #results = response['results']
    
    try:
    # run a request using our params dictionary
      response = requests.get(base_url, params=params)
            #append to hotel data frame
      #  pprint(results)
      print(len(response.content))
      if (len(response.content)>100):
        response1 = response.json()

        for column in source_df.columns[5:]:
            source_df[column].iloc[index]=response1['series']['seriesInfo'][0]['values'][column]
          
      
    #   source_df['Domain_id'].iloc[i]=response[0]['ids'][0]['id']
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")

source_df.to_csv('../VIC1.csv',index=False)
```
