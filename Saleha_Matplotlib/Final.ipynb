```python
plt.figure(figsize=(15,10))
sns.barplot(data=df,x='bedrooms',y='price',palette='Set2')
plt.xlabel("Number of Bedroom", size=20)
plt.ylabel("price", size=20)
plt.title("Total Price of Houses in Melbourne by Bedroom", size=25)
plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=df,x='bathrooms',y='price',palette='Set2')
plt.xlabel("Number of Bathroom", size=20)
plt.ylabel("price", size=20)
plt.title("Total Price of Houses in Melbourne by Bathroom", size=25)
plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=df,x='carspaces',y='price',palette='Set2')
plt.xlabel("Number of Carport", size=20)
plt.ylabel("price", size=20)
plt.title("Total Price of Houses in Melbourne by Carport", size=25)
plt.tight_layout()
```

```python
# plotting a heat map for all numeric variables
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),cmap = 'coolwarm',linewidth = 1,annot= True, annot_kws={"size": 9})
plt.title('HEAT MAP', fontsize=15)
plt.show()
```
